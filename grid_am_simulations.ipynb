Create an animated GIF for displaying simulations of contracting microtubule networks with varying shapes.

```python
import os
import glob
import tqdm
import matplotlib.pyplot as plt
from skimage.io import imread
from helper_functions import *
```

```python
data_dir = os.path.join(os.getcwd(), 'sim_data_112x112/') 
list_patterns = [
    16312, 49112, 65380, 10047, 40955, 27367, 28391, 20108,
    25699, 62200, 30159, 64876, 59942, 53216, 7967,  53241,
    1022,  11953, 57296, 56300, 25383, 11838, 61948, 612,
    18261, 61223, 31464, 8983,  30272, 26486, 11232, 5109,
    64118, 51698, 62063, 28600, 2024,  49119, 20458, 4883,
    831,   14135, 8174,  64032, 62959, 5958,  64125, 32460
]
list_dirs = [data_dir + 'pattern_' + str(pattern).zfill(5) for pattern in list_patterns]
```

```python
figures = []
for i in tqdm.tqdm(range(0,170,5)):
    paths_frames = [sorted(glob.glob(dir_ + '/*'))[i] for  dir_ in list_dirs]
    n_rows = 6
    n_columns = 8
    total_n = n_rows * n_columns
    c = 0
    f = plt.figure(dpi = 150)
    for path in paths_frames:
        im = imread(path)
        plt.subplot(n_rows, n_columns, c+1)
        plt.imshow(im, cmap = 'bone')
        plt.xticks([], [])
        plt.yticks([], [])
        c +=1
    plt.subplots_adjust(hspace=0.2, wspace=0)
    figures.append(matplotlib_fig_to_np_array(f))
    plt.close()
embed_images(figures, duration = .05)
```
